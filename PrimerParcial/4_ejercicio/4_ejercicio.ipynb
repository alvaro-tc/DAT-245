```python
%%writefile ejercicio.py
from kanren import Relation, facts, run, var, eq

# Definir relaciones
padre = Relation()
madre = Relation()
hermano = Relation()
primo = Relation()
tio = Relation()
abuelo = Relation()

# Establecer relaciones con facts
facts(padre, ("Mario", "Alvaro"),
             ("Mario", "Adriana"))
facts(madre, ("Susana", "Alvaro"),
             ("Susana", "Adriana"))

facts(hermano, ("Adriana", "Alvaro"),
               ("Alvaro", "Adriana"),
               ("Rene", "Juan"),
               ("Susana", "Juan"),
               ("Rene", "Susana"),
               ("Juan", "Rene"))

facts(abuelo, ("Genara Cupiticona", "Alvaro"),
               ("Jaime Torrez", "Alvaro"),
               ("Hernan Calle", "Alvaro"),
               ("Maxima Chambi", "Alvaro"),
               ("Genara Cupiticona", "Adriana"),
               ("Jaime Torrez", "Adriana"),
               ("Hernan Calle", "Adriana"),
               ("Maxima Chambi", "Adriana"),
               ("Hernan Calle", "Maite"),
               ("Maxima Chambi", "Maite"),
               ("Hernan Calle", "Zuelem"),
               ("Maxima Chambi", "Zuelem"))

facts(tio, ("Rene", "Alvaro"),
           ("Milena Limachi", "Alvaro"),
           ("Victor", "Alvaro"),
           ("Rene", "Adriana"),
           ("Milena Limachi", "Adriana"),
           ("Victor", "Adriana"),
           ("Mario", "Maite"),
           ("Susana", "Maite"),
           ("Mario", "Zuelem"),
           ("Susana", "Zuelem"))
facts(primo, ("Maite", "Alvaro"),
    ("Zuelem", "Alvaro"),
    ("Maite", "Adriana"),
    ("Zuelem", "Adriana"),
    ("Alvaro", "Maite"),
    ("Alvaro", "Zuelem"),
    ("Adriana", "Maite"),
    ("Adriana", "Zuelem")
)

# Consultar quiénes son los primos de "Alvaro"
x = var()
result = run(10, x, primo(x, "Alvaro"))
print("Primos de Alvaro:", result)
```
